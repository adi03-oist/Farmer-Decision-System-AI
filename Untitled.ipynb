```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Agriculture_price_dataset.csv")
df.head()
```

```python
df["Price Date"] = pd.to_datetime(df["Price Date"])
df = df.sort_values("Price Date")
```

```python
vegetables = ["Tomato","Onion","Potato","Cabbage","Green Chilli","Garlic","Peas"]
```

```python
veg_df = df[df["Commodity"].isin(vegetables)]
```

```python
veg_df = veg_df.copy()
veg_df["day"] = veg_df["Price Date"].dt.dayofyear
veg_df["month"] = veg_df["Price Date"].dt.month
```

```python
import matplotlib.pyplot as plt

for veg in ["Tomato","Onion","Potato"]:
    temp = veg_df[veg_df["Commodity"]==veg].copy()
    
    X = temp[["day","month"]]
    temp["Predicted"] = models[veg].predict(X)

    plt.figure(figsize=(10,4))
    plt.plot(temp["Price Date"], temp["Modal_Price"], label="Actual Price")
    plt.plot(temp["Price Date"], temp["Predicted"], label="AI Prediction")
    
    plt.legend()
    plt.title(f"{veg} Price: Actual vs AI Prediction")
    plt.xlabel("Date")
    plt.ylabel("Price (Rs)")
    
    plt.savefig(f"{veg.lower()}_ai.png")
    plt.show()
```

```python
plt.figure(figsize=(12,6))

for veg in valid_vegs:
    temp = veg_df[veg_df["Commodity"]==veg]
    temp = temp.sort_values("Price Date")
    temp["MA"] = temp["Modal_Price"].rolling(30).mean()   # 30-day average

    plt.plot(temp["Price Date"], temp["MA"], label=veg)

plt.legend()
plt.title("30-Day Moving Average Vegetable Price Trends")
plt.xlabel("Date")
plt.ylabel("Price (Rs)")
plt.grid(True)
plt.savefig("30day_moving_avg.png")
plt.show()
```
